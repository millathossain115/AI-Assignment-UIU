```python
path = "/content/diabetes2.csv"
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(path)

display(df)
```

```python
npArray=df.to_numpy()
display(npArray)
```

```python
print(type(npArray))
print(npArray)
```

```python
from sklearn import preprocessing
normArray = preprocessing.normalize(npArray)
print(normArray)
```

```python
import random
TrainSet = []
TestSet =[]

random.shuffle(normArray)

for i in range(len(normArray)):
  R = random.random()
  # print(R)
  if R>=0 and R<=0.8:
    TrainSet.append(normArray[i])
  else:
    TestSet.append(normArray[i])

print("Train Set........................................")
for i in range(len(TrainSet)):
   print(TrainSet[i])


```

```python
print("Test Set........................................")
for i in range(len(TestSet)):
   print(TestSet[i])
```

```python
Glucose = []
BloodPressure = []
SkinThickness = []
Insulin = []
BMI = []
DiabetesPer = []
Age = []
Outcome = []



for i in range(len(TrainSet)):
  Glucose.append(TrainSet[i][0])
  BloodPressure.append(TrainSet[i][1])
  SkinThickness.append(TrainSet[i][2])
  Insulin.append(TrainSet[i][3])
  BMI.append(TrainSet[i][4])
  DiabetesPer.append(TrainSet[i][5])
  Age.append(TrainSet[i][6])
  Outcome.append(TrainSet[i][7])

def avg(arr):
  s = sum(arr)
  a = s/len(arr)
  print(a)

avg(Glucose)
avg(BloodPressure)
avg(SkinThickness)
avg(Insulin)
avg(BMI)
avg(DiabetesPer)
avg(Age)
avg(Outcome)

    
```

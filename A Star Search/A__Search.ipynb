```python
def aStarAlgo(start_node, stop_node):

    open_set = set(start_node)
    closed_set = set()
    g = {}  # store distance from starting node
    parents = {}  # parents contains an adjacency map of all nodes

    # ditance of starting node from itself is zero
    g[start_node] = 0
    # start_node is root node i.e it has no parent nodes
    # so start_node is set to its own parent node
    parents[start_node] = start_node

    while len(open_set) > 0:
        n = None

        # node with lowest f() is found
        for v in open_set:
            if n == None or g[v] + heuristic(v) < g[n] + heuristic(n):
                n = v

        if n == stop_node or Graph_nodes[n] == None:
            pass
        else:
            for (m, weight) in get_neighbors(n):
                # nodes 'm' not in first and last set are added to first
                # n is set its parent
                if m not in open_set and m not in closed_set:
                    open_set.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight

                # for each node m,compare its distance from start i.e g(m) to the
                # from start through n node
                else:
                    if g[m] > g[n] + weight:
                        # update g(m)
                        g[m] = g[n] + weight
                        # change parent of m to n
                        parents[m] = n

                        # if m in closed set,remove and add to open
                        if m in closed_set:
                            closed_set.remove(m)
                            open_set.add(m)

        if n == None:
            print('Path does not exist!')
            return None

        # if the current node is the stop_node
        # then we begin reconstructin the path from it to the start_node
        if n == stop_node:
            path = []

            while parents[n] != n:
                path.append(n)
                n = parents[n]

            path.append(start_node)

            path.reverse()

            print('Path: ', end='')
            for x in range(len(path)):
                if not x == len(path)-1:
                    print(path[x], end='-->')
                else:
                    print(path[x])

            return path
        # remove n from the open_list, and add it to closed_list
        # because all of his neighbors were inspected
        open_set.remove(n)
        closed_set.add(n)
    print('Path does not exist!')
    return None

# define fuction to return neighbor and its distance
# from the passed node


def get_neighbors(v):
    if v in Graph_nodes:
        return Graph_nodes[v]
    else:
        return None

# function returns heuristic distance for all nodes


def heuristic(n):
    # get_heuristic_distances() function
    # H_dist = get_heuristic_distances()
    H_dist = {
        '0': 14,
        '1': 12,
        '2': 11,
        '3': 6,
        '4': 4,
        '5': 11,
        '6': 0

    }

    return H_dist[n]


# function to take Graph_nodes from the user
def get_graph_nodes():
    Graph_nodes = {}
    print('Enter the number of nodes in the graph: ')
    num_nodes = int(input())
    for i in range(num_nodes):
        print('Enter the node name: ')
        node_name = input()
        print('Enter the number of neighbors for node ' + node_name + ': ')
        num_neighbors = int(input())
        neighbors = []
        for j in range(num_neighbors):
            print('Enter the neighbor name: ')
            neighbor_name = input()
            print('Enter the distance between ' +
                  node_name + ' and ' + neighbor_name + ': ')
            distance = int(input())
            neighbors.append((neighbor_name, distance))
        Graph_nodes[node_name] = neighbors
    return Graph_nodes


# write a function to take heuristic distances from the user
def get_heuristic_distances():
    H_dist = {}
    print('Enter the number of nodes in the graph: ')
    num_nodes = int(input())
    for i in range(num_nodes):
        print('Enter the node name: ')
        node_name = input()
        print('Enter the heuristic distance for node ' + node_name + ': ')
        distance = int(input())
        H_dist[node_name] = distance
    return H_dist

# Graph_nodes = get_graph_nodes()


Graph_nodes = {
    '0': [('1', 4), ('2', 3)],
    '1': [('4', 12), ('5', 5)],
    '2': [('3', 7), ('4', 10)],
    '3': [('4', 2)],
    '4': [('6', 5)],
    '5': [('6', 16)],
    '6': []

}

path = aStarAlgo('0', '6')

# total distance of the path


def get_path_distance(path):
    distance = 0
    for i in range(len(path)-1):
        distance += get_distance(path[i], path[i+1])
    return distance
# get_distance function


def get_distance(node1, node2):
    if node1 in Graph_nodes and node2 in Graph_nodes:
        for (n, weight) in Graph_nodes[node1]:
            if n == node2:
                return weight
    return None


# the function to get the distance
print('Totat Actual Cost:', get_path_distance(path))
```
